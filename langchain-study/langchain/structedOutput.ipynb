```python
from pydantic import BaseModel, Field
from dotenv import load_dotenv

load_dotenv()

from langchain_openai import ChatOpenAI

model = ChatOpenAI(model="Qwen/Qwen3-8B", base_url="https://api.siliconflow.cn")

from langchain.agents import create_agent, AgentState
```

## 默认走提供商内置策略

```python
class ContactInfo(BaseModel):
    """人物联系方式"""
    name: str = Field(description="人物的名字")
    email: str = Field(description="人物的邮件地址")
    phone: str = Field(description="人物的电话号码")


agent = create_agent(
    model,
    response_format=ContactInfo
)

structured_response = agent.invoke({"messages": [{"role": "user",
                                                  "content": "请帮我提取人物的联系方式:林凯淼的邮箱是3027383@qq.com,经常使用的号码是17867722311,他很喜欢玩游戏和听歌"}]})

print(structured_response['structured_response'])
```

## 工具结构化输出,还可以配置Handle_error进行自定义的错误处理

```python
from langchain.agents import create_agent
from langchain.agents.structured_output import ToolStrategy

agent = create_agent(
    model,
    #tool_message_content是返回给ai工具调用的结果
    response_format=ToolStrategy(ContactInfo,tool_message_content="提取了人物联系方式"),
)

result = agent.invoke({
    "messages": [{"role": "user",
                  "content": "请帮我提取人物的联系方式:林凯淼的邮箱是3027383@qq.com,经常使用的号码是17867722311,他很喜欢玩游戏和听歌"}]
})
print(result["structured_response"])
```
